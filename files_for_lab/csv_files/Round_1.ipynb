```python
import pandas as pd
```

```python
file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')
file3 = pd.read_csv('file3.csv')
```

```python
print(file1.shape)
print(file2.shape)
print(file3.shape)
```

```python
file1.columns
```

```python
file2.columns
```

```python
file3.columns
```

```python
cols = []
for i in range(len(file1.columns)):
    cols.append(file1.columns[i].lower())
file1.columns = cols

file1.rename(columns = {'st':'state'}, inplace = True)

file1.columns
```

```python
cols = []
for i in range(len(file2.columns)):
    cols.append(file2.columns[i].lower())
file2.columns = cols

file2.rename(columns = {'st':'state'}, inplace = True)

file2 = file2[['customer', 'state', 'customer lifetime value', 'education', 'gender',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'total claim amount', 'vehicle class']]

file2.columns
```

```python
cols = []
for i in range(len(file3.columns)):
    cols.append(file3.columns[i].lower())
file3.columns = cols

file3.columns
```

```python
data = pd.concat([file1, file2, file3], axis=0)

data
```

```python
data.dtypes

# Numerical: All columns labeled 'float64'

# Categorical: All columns labeled 'object'
```

```python
data = data.drop(['gender', 'number of open complaints'], axis=1)
```

```python
data.head()
```

```python
data['customer lifetime value'] = data['customer lifetime value'].str.rstrip('%').astype('float') / 100.0

print(data['customer lifetime value'].dtypes)

print(data.head())

print(data.shape)
```

```python
data = data.drop_duplicates()

data.shape
```

```python
data_with_income = data[data['income']>0]

data_with_income
```
